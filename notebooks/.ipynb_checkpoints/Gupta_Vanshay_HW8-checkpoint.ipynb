```python
for i in range(30):
    df_m_test=df_m.sample(frac=0.2)
    df_m_train = df_m.drop(df_m_test.index)
    df_b_test=df_b.sample(frac=0.2)
    df_b_train = df_b.drop(df_b_test.index)
    train=pd.concat([df_m_train, df_b_train])
    test=pd.concat([df_m_test, df_b_test])
    
    X_train=train.drop([1], axis=1)
    y_train=train[1]
    X_test=test.drop([1], axis=1)
    y_test=test[1]
    X_train=X_train.reset_index(drop=True)
    y_train=y_train.reset_index(drop=True)
    
    scaler = MinMaxScaler()
    scaler.fit(X_train)
    X_train=scaler.transform(X_train)
    X_test=scaler.transform(X_test)

    model=KMeans(n_clusters=2, init='k-means++', n_init=5, random_state=random.randint(0,100))
    model.fit(X_train)
    n_neighbors = NearestNeighbors(n_neighbors=30).fit(X_train)
    distances, indices = n_neighbors.kneighbors(model.cluster_centers_)
    cluster_0= y_train.loc[indices[0]]
    majority_polling_0=cluster_0.value_counts()
    cluster_1=y_train.loc[indices[1]]
    majority_polling_1=cluster_1.value_counts()

    pred = model.labels_
    pred=pd.DataFrame(pred)
    pred_0=pred[pred[0]==0]
    pred_1=pred[pred[0]==1]
    max_index_0=np.argmax(majority_polling_0)
    max_index_1=np.argmax(majority_polling_1)

    pred_0['class']=max_index_0
    pred_1['class']=max_index_1
    y_pred_train=pd.concat([pred_0['class'],pred_1['class']],axis=0)
    y_pred_train=y_pred_train.sort_index()
    
    y_pred = model.predict(X_test)
    
    precision.append(precision_score(y_test, y_pred))
    recall.append(recall_score(y_test, y_pred))
    accuracy.append(accuracy_score(y_test, y_pred)) 
    fscore.append(f1_score(y_test, y_pred))
    fpr, tpr, _ = roc_curve(y_test, y_pred)
    aucl.append(auc(fpr, tpr))

    precision_train.append(precision_score(y_train, y_pred_train))
    recall_train.append(recall_score(y_train, y_pred_train))
    accuracy_train.append(accuracy_score(y_train, y_pred_train)) 
    fscore_train.append(f1_score(y_train, y_pred_train))
    fpr_train, tpr_train, _ = roc_curve(y_train, y_pred_train)
    auc_train.append(auc(fpr_train, tpr_train))
```
